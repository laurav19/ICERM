# Cyclic evasion

```python
%pylab inline
from plots import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

### 4-bugs case

```python
def evasion_01(state,time):
    
    state = state.reshape([4,2])
    
    g0 = (state[0]-state[1])/np.linalg.norm(state[0]-state[1])
    g1 = (state[1]-state[2])/np.linalg.norm(state[1]-state[2])
    g2 = (state[2]-state[3])/np.linalg.norm(state[2]-state[3])
    g3 = (state[3]-state[0])/np.linalg.norm(state[3]-state[0])
    
    g = np.array([g0,g1,g2,g3])
    
    return g.reshape([8])
```

```python
def evasion_02(state,time):
    
    n = state.shape[0]/2
    state = np.append(state, [state[0], state[1]])
    state = state.reshape([n+1,2])
    
    y = np.array([])
    
    for i in xrange (0,n):
        g = (state[i]-state[i+1])/np.linalg.norm(state[i]-state[i+1])
        y = np.append(y,g)
    
    return y
```

```python
def plot_polygon(pol):
    
    n = pol.shape[0]
    dots = np.reshape(np.append(pol[:,:],pol[0,:]), [n+1,2])
    
    plt.figure(1, figsize=(3,3))
    plt.plot(dots[:,0],dots[:,1],lw=5)
    plt.grid()
    axis('equal');
```

```python
def plot_solution(sol):
    
    steps = sol.shape[0]
    n = sol.shape[1]
    plt.figure(1, figsize=(8,8))
    
    initial = np.reshape(np.append(solution[0,:,:],solution[0,0,:]), [n+1,2])
    final = np.reshape(np.append(solution[steps-1,:,:],solution[steps-1,0,:]), [n+1,2])
    for i in xrange(0,n):
        bug = sol[:,i,:]
        plt.plot(bug[:,0], bug[:,1])
    
    plt.plot(initial[:,0],initial[:,1])
    plt.plot(final[:,0],final[:,1])
    axis('equal');
```

```python
steps = 1000
time_limit = 10000
time = np.linspace(0,time_limit,steps)
```

#### Square

```python
n = 4
```

```python
y_01 = np.zeros((n,2))
y_01[0]=[0,0]
y_01[1]=[0,1]
y_01[2]=[1,1]
y_01[3]=[1,0]

plot_polygon(y_01)
```

```python
solution = odeint(func=evasion_02, y0 = y_01.reshape([2*n]), t = time)
solution = solution.reshape((steps,n,2))
```

```python
plot_solution(solution)
```

#### Parallelogram

```python
y_02 = np.zeros((n,2))
y_02[0]=[-1,0]
y_02[1]=[0,1]
y_02[2]=[1,1]
y_02[3]=[0,0]

plot_polygon(y_02)
```

```python
solution = odeint(func=evasion_02, y0 = y_02.reshape([2*n]), t = time)
solution = solution.reshape((steps,n,2))
```

```python
plot_solution(solution)
```

#### Trapezoid

```python
steps = 1000
time_limit = 1e11
time = np.linspace(0,time_limit,steps)
```

```python
eps = 0.1

y_03 = np.zeros((4,2))
y_03[0]=[0-eps,0]
y_03[1]=[0,1]
y_03[2]=[1,1]
y_03[3]=[1+eps,0]

plot_polygon(y_03)
```

```python
solution = odeint(func=evasion_02, y0 = y_03.reshape([2*n]), t = time)
solution = solution.reshape((steps,n,2))
```

```python
plot_solution(solution)
```
