# Cyclic evasion

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

### 4-bugs case

```python
def evasion_01(state,time):
    
    state = state.reshape([4,2])
    
    g0 = (state[0]-state[1])/np.linalg.norm(state[0]-state[1])
    g1 = (state[1]-state[2])/np.linalg.norm(state[1]-state[2])
    g2 = (state[2]-state[3])/np.linalg.norm(state[2]-state[3])
    g3 = (state[3]-state[0])/np.linalg.norm(state[3]-state[0])
    
    g = np.array([g0,g1,g2,g3])
    
    return g.reshape([8])
```

```python
steps = 100
time_limit = 10
time = np.linspace(0,time_limit,steps)
```

#### Square

```python
y_01 = np.zeros((4,2))
y_01[0]=[0,0]
y_01[1]=[0,1]
y_01[2]=[1,1]
y_01[3]=[1,0]

initial = np.array([y_01[0],y_01[1],y_01[2],y_01[3], y_01[0]])
plt.figure(1, figsize=(3,3))
plt.plot(initial[:,0],initial[:,1])
plt.grid()
axis('equal')
plt.xlim(-.5,1.5);
```

```python
y_01 = y_01.reshape([8])
```

```python
solution = odeint(func=evasion_01, y0 = y_01, t = time)
solution = solution.reshape((steps,4,2))
```

```python
bug1 = solution[:,0,:]
bug2 = solution[:,1,:]
bug3 = solution[:,2,:]
bug4 = solution[:,3,:]
final = np.array([bug1[steps-1],bug2[steps-1],bug3[steps-1],bug4[steps-1], bug1[steps-1]])
```

```python
plt.figure(1, figsize=(8,8))
plt.plot(bug1[:,0],bug1[:,1])
plt.plot(bug2[:,0],bug2[:,1])
plt.plot(bug3[:,0],bug3[:,1])
plt.plot(bug4[:,0],bug4[:,1])
plt.plot(initial[:,0],initial[:,1])
plt.plot(final[:,0],final[:,1])
axis('equal');
```
