```python
%pylab inline
from plots import *
from matplotlib import animation
from scipy.integrate import odeint
```

```python
from tempfile import NamedTemporaryFile

VIDEO_TAG = """<video controls>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Tu navegador no soporta este formato de vídeo.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, writer='avconv', fps=20, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    
    return VIDEO_TAG.format(anim._encoded_video)
```

```python
from IPython.display import HTML

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))
```

```python
n = 4
```

```python
y_02 = np.zeros((n,2))
y_02[0]=[-0.5,0]
y_02[1]=[0,1]
y_02[2]=[0.5,0]
y_02[3]=[0,-1]

plot_polygon(y_02)
```

```python
def evasion_02(state,time):
    
    n = state.shape[0]/2
    state = np.append(state, [state[0], state[1]])
    state = state.reshape([n+1,2])
    
    y = np.array([])
    
    for i in xrange (0,n):
        g = (state[i]-state[i+1])/np.linalg.norm(state[i]-state[i+1])
        y = np.append(y,g)
    
    return y
```

```python
steps = 300
time_limit = 1e10
time = np.linspace(0,time_limit,steps)
```

```python
solution = odeint(func=evasion_02, y0 = y_02.reshape([2*n]), t = time)
solution = solution.reshape((steps,n,2))
```

```python
plot_solution(solution)
```

```python
# Figura, ejes y el elemento plot que vamos a animar
fig = plt.figure(figsize=(8,8))
ax = plt.axes(xlim=(limmin, limmax), ylim=(limmin, limmax))
ax.set_aspect('equal')
initial, = ax.plot([], [], lw=2)
line1, = ax.plot([], [], lw=2)
line2, = ax.plot([], [], lw=2)
line3, = ax.plot([], [], lw=2)
line4, = ax.plot([], [], lw=2)
final, = ax.plot([], [], lw=2)

# Dibujamos el fondo de cada fragmento
def init():
    initial.set_data([],[])
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    line4.set_data([], [])
    final.set_data([],[])
    return initial, line1, line2, line3, line4, final,

# Función de animación, se llama secuencialmente
def animate(i):
    
    x = np.reshape(np.append(solution[0,:,:],solution[0,0,:]), [n+1,2])
    x = x[:,0]
    y = np.reshape(np.append(solution[0,:,:],solution[0,0,:]), [n+1,2])
    y = y[:,1]
    initial.set_data(x,y)
    
    x = np.reshape(np.append(solution[i,:,:],solution[i,0,:]), [n+1,2])
    x = x[:,0]
    y = np.reshape(np.append(solution[i,:,:],solution[i,0,:]), [n+1,2])
    y = y[:,1]
    final.set_data(x,y)
    
    bug = solution[0:i+1,0,:]
    x = bug[:,0]
    y = bug[:,1]
    line1.set_data(x,y)
    
    bug = solution[0:i+1,1,:]
    x = bug[:,0]
    y = bug[:,1]
    line2.set_data(x,y)
    
    bug = solution[0:i+1,2,:]
    x = bug[:,0]
    y = bug[:,1]
    line3.set_data(x,y)
    
    bug = solution[0:i+1,3,:]
    x = bug[:,0]
    y = bug[:,1]
    line4.set_data(x,y)
    return initial, line1, line2, line3, line4, final,
```

```python
# Para detalles de los parámetros de la funcion: http://matplotlib.org/api/animation_api.html?highlight=funcanimation
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=steps, interval=20, blit=True)
```

```python
def center_mass(pol):
    n = pol.shape[0]
    cx = sum(pol[:,0])
    cy = sum(pol[:,1])
    c = np.array([cx, cy])
    c = c/n
    return pol-c
```
