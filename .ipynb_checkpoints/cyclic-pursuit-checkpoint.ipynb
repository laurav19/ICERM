# Cyclic pursuit

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def pursuit_01(state,time):
    
    n = state.shape[0]/2
    state = np.append(state, [state[0], state[1]])
    state = state.reshape([n+1,2])
    
    y = np.array([])
    
    for i in xrange (0,n):
        g = (state[i+1]-state[i])/np.linalg.norm(state[i+1]-state[i])
        y = np.append(y,g)
    
    return y
```

```python
def pursuit_02(state,time):
    
    n = state.shape[0]/2
    state = np.append(state, [state[0], state[1], state[2], state[3]])
    state = state.reshape([n+2,2])
    
    y = np.array([])
    
    for i in xrange (0,n):
        g = (state[i+2]-state[i])/np.linalg.norm(state[i+2]-state[i])
        y = np.append(y,g)
    
    return y
```

```python
steps = 100
time_limit = 1
time = np.linspace(0,time_limit,steps)
```

```python
y_02 = np.zeros((4,2))
y_02[0]=[-1,0]
y_02[1]=[0,1]
y_02[2]=[1,1]
y_02[3]=[0,0]

initial = np.array([y_02[0],y_02[1],y_02[2],y_02[3], y_02[0]])
plt.figure(1, figsize=(3,3))
plt.plot(initial[:,0],initial[:,1])
plt.grid()
axis('equal');
```

```python
y_02 = y_02.reshape([8])
```

```python
solution = odeint(func=pursuit_01, y0 = y_02, t = time)
solution = solution.reshape((steps,4,2))
```

```python
bug1 = solution[:,0,:]
bug2 = solution[:,1,:]
bug3 = solution[:,2,:]
bug4 = solution[:,3,:]
final = np.array([bug1[steps-1],bug2[steps-1],bug3[steps-1],bug4[steps-1], bug1[steps-1]])
```

```python
plt.figure(1, figsize=(8,8))
plt.plot(bug1[:,0],bug1[:,1])
plt.plot(bug2[:,0],bug2[:,1])
plt.plot(bug3[:,0],bug3[:,1])
plt.plot(bug4[:,0],bug4[:,1])
plt.plot(initial[:,0],initial[:,1])
plt.plot(final[:,0],final[:,1])
axis('equal');
```

```python
steps = 100
time_limit = 15
time = np.linspace(0,time_limit,steps)
```

```python
y_02 = np.zeros((5,2))
y_02[0]=[0,1]
y_02[1]=[.587,.809]
y_02[2]=[9.51,3.09]
y_02[3]=[-9.51,3.09]
y_02[4]=[-.587,.809]

initial = np.array([y_02[0],y_02[1],y_02[2],y_02[3], y_02[4], y_02[0]])
plt.figure(1, figsize=(3,3))
plt.plot(initial[:,0],initial[:,1])
plt.grid()
axis('equal');
```

```python
bug1 = solution[:,0,:]
bug2 = solution[:,1,:]
bug3 = solution[:,2,:]
bug4 = solution[:,3,:]
bug5 = solution[:,4,:]
final = np.array([bug1[steps-1],bug2[steps-1],bug3[steps-1],bug4[steps-1],bug5[steps-1], bug1[steps-1]])
```

```python
plt.figure(1, figsize=(8,8))
plt.plot(bug1[:,0],bug1[:,1])
plt.plot(bug2[:,0],bug2[:,1])
plt.plot(bug3[:,0],bug3[:,1])
plt.plot(bug4[:,0],bug4[:,1])
plt.plot(bug5[:,0],bug5[:,1])
plt.plot(initial[:,0],initial[:,1])
plt.plot(final[:,0],final[:,1])
axis('equal');
```
