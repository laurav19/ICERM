# Cyclic evasion

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
N = 100
time_limit = 10
time = np.linspace(0,time_limit,N)
```

```python
y01 = np.zeros((4,2))
```

```python
y01[0]=[0,0]
y01[1]=[0,1]
y01[2]=[1,1]
y01[3]=[1,0]
y01
```

```python
def evasion_02(state,time):
    
    state = state.reshape([4,2])
    
    g0 = (state[0]-state[1])/np.linalg.norm(state[0]-state[1])
    g1 = (state[1]-state[2])/np.linalg.norm(state[1]-state[2])
    g2 = (state[2]-state[3])/np.linalg.norm(state[2]-state[3])
    g3 = (state[3]-state[0])/np.linalg.norm(state[3]-state[0])
    
    g = np.array([g0,g1,g2,g3])
    
    return g.reshape([8])
```

```python
y02 = y01.reshape([8])
y02
```

```python
evasion_02(y02,0)
```

```python
solution = odeint(func=evasion_02, y0 = y02, t = time)
```

```python
solution = solution.reshape((N,4,2))
solution
```

```python
bug1 = solution[:,0,:]
bug2 = solution[:,1,:]
bug3 = solution[:,2,:]
bug4 = solution[:,3,:]
initial = np.array([bug1[0],bug2[0],bug3[0],bug4[0], bug1[0]])
final = np.array([bug1[N-1],bug2[N-1],bug3[N-1],bug4[N-1], bug1[N-1]])
```

```python
plt.figure(1, figsize=(8,8))
plt.plot(bug1[:,0],bug1[:,1])
plt.plot(bug2[:,0],bug2[:,1])
plt.plot(bug3[:,0],bug3[:,1])
plt.plot(bug4[:,0],bug4[:,1])
plt.plot(initial[:,0],initial[:,1])
plt.plot(final[:,0],final[:,1]);
```
